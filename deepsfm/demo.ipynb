```python
from pathlib import Path
from image_registration import DeepClipFeatureMatchingCOLMAP
import pycolmap
from lightglue import LightGlue, SuperPoint, DISK, SIFT, ALIKED
import cv2
from functools import partial
```

```python
extractor = DISK(max_num_keypoints=2048).eval()
matcher = LightGlue(features='disk').eval()
```

```python
# !pip install yt-dlp
# !yt-dlp -f mp4 https://www.youtube.com/shorts/goB7lwFEAl8 -o video.mp4
```

```python
options = pycolmap.ExhaustiveMatchingOptions()

pair_generator = partial(pycolmap.ExhaustivePairGenerator, options=options)
```

```python
vid_path = "video.mp4"
vid = cv2.VideoCapture(vid_path)
total_frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
fps = vid.get(cv2.CAP_PROP_FPS)
duration = int(total_frames / fps)

registration = DeepClipFeatureMatchingCOLMAP(
    clips=[[0, duration]],
    vid_path=vid_path,
    feature_extractor=extractor,
    feature_matcher=matcher,
    pairgenerator=pair_generator,
    device="cpu",
    sample_rate=0.1,
)
```

```python
recon = registration.register()
```
